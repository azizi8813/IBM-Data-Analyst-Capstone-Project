## Demo: How to work with database


Download database file.


```python
import requests
urln='https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m4_survey_data.sqlite'
resp=requests.get(urln)
with open('m4_survey_data.sqlite','wb') as f:
    f.write(resp.content)
    print("Database file downloaded successfully.")
```

Connect to the database.


```python
import sqlite3
conn = sqlite3.connect("m4_survey_data.sqlite") # open a database connection
```

Import pandas module.


```python
import pandas as pd
```

## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
```

```python
Query='''select * from master'''
df=pd.read_sql_query(Query,conn)
```

```python
sns.histplot(data=df,x='ConvertedComp',bins=120)
formatter = mticker.FuncFormatter(lambda x, pos: f"{x:,.0f}")

# Apply the formatter to the x-axis ticks
plt.gca().xaxis.set_major_formatter(formatter)
plt.xticks(rotation=75)
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
sns.boxplot(data=df,x='Age')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
sns.scatterplot(data=df,x='Age',y='WorkWeekHrs')
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
sns.scatterplot(data=df,y='WorkWeekHrs',x='CodeRevHrs',size='Age')
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
Query='''select * from DatabaseDesireNextYear'''
df=pd.read_sql_query(Query,conn)
df.head()
```

```python
df1=df.groupby('DatabaseDesireNextYear',as_index=False).count()
df2=df1.sort_values(by='Respondent',ascending=False).head()
sns.set_theme()  # Set Seaborn's default theme
plt.figure(figsize=(8, 8))
plt.pie(x=df2['Respondent'],labels=df2['DatabaseDesireNextYear'],autopct='%1.1f%%')
plt.title('Pie Chart')

plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
Query='''select * from master'''
df=pd.read_sql_query(Query,conn)
```

```python
df2=df[(df['Age']>=30)&(df['Age']<=35)]
df3=df2.groupby('Age')[['WorkWeekHrs','CodeRevHrs']].median()
df3.plot(kind='bar', stacked=True)
plt.xticks(rotation=0)
plt.legend(loc='lower right')
plt.show()
```

```python
df2=df[(df['Age']>=30)&(df['Age']<=35)]
df3=df2.groupby('Age')[['WorkWeekHrs','CodeRevHrs']].mean()
df3.plot(kind='bar', stacked=True)
plt.xticks(rotation=0)
plt.legend(loc='lower right')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
df2=df[(df['Age']>=45)&(df['Age']<=60)]
df3=df2.groupby('Age',as_index=False)['ConvertedComp'].median()
sns.lineplot(data=df3,x='Age',y='ConvertedComp')
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
mainbranch_counts = df['MainBranch'].value_counts()
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 6))  
sns.barplot(x=mainbranch_counts.values, y=mainbranch_counts.index, palette='viridis',hue=mainbranch_counts.index,legend=False)

plt.xlabel('Count')
plt.ylabel('Main Branch')
plt.title('Horizontal Bar Chart of MainBranch')
plt.show()
```
