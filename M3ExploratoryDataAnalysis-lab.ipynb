## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
sns.boxplot(data=df,x='ConvertedComp')
formatter = mticker.FuncFormatter(lambda x, pos: f"{x:,.0f}")

# Apply the formatter to the x-axis ticks
plt.gca().xaxis.set_major_formatter(formatter)
plt.xticks(rotation=75)
plt.show()
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
q1 = df['ConvertedComp'].quantile(0.25)
q3 = df['ConvertedComp'].quantile(0.75)
iqr = q3-q1
q1, q3, iqr
```

Find out the upper and lower bounds.


```python
Upper_bound= q3 + 1.5 * iqr
Lower_bound= q1 - 1.5 * iqr
print('Upper bound : {}'.format(Upper_bound))
print('Lower bound : {}'.format(Lower_bound))
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
df[(df['ConvertedComp'] > Upper_bound) | (df['ConvertedComp']<Lower_bound)].shape[0]
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
df1=df[(df['ConvertedComp'] > Upper_bound) | (df['ConvertedComp']<Lower_bound)].index
dfn=df.drop(index=df1,axis=0)
df.head()
```

```python
dfn.shape
```

```python
df_filtered = df[(df['ConvertedComp'] >= Lower_bound) & (df['ConvertedComp'] <= Upper_bound)]
df_filtered.shape
```
